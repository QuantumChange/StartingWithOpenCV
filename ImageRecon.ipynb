# Image Recognition For Beginner

## Image Loading 

```python
import cv2

image_path='D:\pyplib\imageRecon\check\Golden.jpg'

image=cv2.imread(image_path)

if image is not None:
    print("Image Loaded successfully")
else:
    print("Failed to Load")
    

print(image.shape)
```

## Image Resize and Flip

```python
resize_img=cv2.resize(image,(200,200))

print(resize_img.shape)

h_img=cv2.flip(image,0)
```

## Image Show with flip

```python
from matplotlib import pyplot as plt
#%matplotlib inline

plt.imshow(cv2.cvtColor(h_img,cv2.COLOR_BGR2RGB))
plt.axis("off")
```

## Image Open in Window

```python
cv2.namedWindow('image',cv2.WINDOW_NORMAL)
cv2.imshow('image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Drawing 2D primitive 

## Drawing Line

```python
import numpy as np

line=cv2.line(image,(0,0),(500,600),(255,0,0),5)

cv2.imshow('line',line)
key=cv2.waitKey()
```

## Drawing Rectangle

```python
rect=cv2.rectangle(image,(20,20),(500,500),(0,255,0),5)

cv2.imshow('rectangle',rect)
key=cv2.waitKey()
```

## Drawing Circle

```python
circ=cv2.circle(image,(500,500),60,(0,0,255),4)

cv2.imshow('circle',circ)
key=cv2.waitKey()
```

## Writing Text 

```python
font=cv2.FONT_HERSHEY_SIMPLEX

hello=cv2.putText(image,'Hello World',(10,500),font,2,(255,255,255),2,cv2.LINE_AA)

cv2.imshow('hello',hello)
key=cv2.waitKey()
```

# Reading Input 

## Reading Input From Key

```python
cv2.namedWindow('image',cv2.WINDOW_NORMAL)
cv2.imshow('image',image)
key=cv2.waitKey()
print(key)
if key==ord('h'):
    h_img=cv2.flip(image,0)
    cv2.imshow('h_img',h_img)
    key=cv2.waitKey()
elif key==ord('h'):
    v_img=cv2.flip(image,1)
    cv2.imshow('v_img',v_img)
    key=cv2.waitKey()
```
